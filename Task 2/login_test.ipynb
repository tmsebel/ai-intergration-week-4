```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import unittest
import os

class LoginTest(unittest.TestCase):
    def setUp(self):
        chrome_options = Options()
        chrome_options.add_argument("--headless") # Run in headless mode
        chrome_options.add_argument("--no-sandbox") # Bypass OS security model
        chrome_options.add_argument("--disable-dev-shm-usage") # Overcome limited resource problems
        # Specify a unique user data directory
        chrome_options.add_argument(f"--user-data-dir={os.path.join(os.getcwd(), 'user_data')}")

        self.driver = webdriver.Chrome(options=chrome_options)
        self.driver.get("https://academy.powerlearnprojectafrica.org/login")  # Replace with actual login URL
        self.success = 0
        self.failure = 0
        self.total_tests = 0

    def test_valid_credentials(self):
        self.total_tests += 1
        driver = self.driver
        # Find elements by their actual IDs on the page if they exist
        try:
            driver.find_element(By.ID, "email").send_keys("valid_user") # Assuming email is used for username
            driver.find_element(By.ID, "password").send_keys("valid_pass123")
            driver.find_element(By.XPATH, "//button[contains(text(), 'Login')]").click() # Assuming login button has text "Login"
            time.sleep(5) # Increased sleep time for page load
            # Check for a success indicator on the page after login
            if "Dashboard" in driver.page_source or "Welcome" in driver.page_source: # Check for dashboard or welcome message
                self.success += 1
            else:
                self.failure += 1
        except Exception as e:
            print(f"Error in valid_credentials test: {e}")
            self.failure += 1


    def test_invalid_credentials(self):
        self.total_tests += 1
        driver = self.driver
        try:
            driver.find_element(By.ID, "email").send_keys("invalid_user") # Assuming email is used for username
            driver.find_element(By.ID, "password").send_keys("wrong_pass")
            driver.find_element(By.XPATH, "//button[contains(text(), 'Login')]").click() # Assuming login button has text "Login"
            time.sleep(5) # Increased sleep time for page load
            # Check for error messages on the page after invalid login
            if "Invalid credentials" in driver.page_source or "Error" in driver.page_source: # Check for specific error messages
                self.success += 1
            else:
                self.failure += 1
        except Exception as e:
            print(f"Error in invalid_credentials test: {e}")
            self.failure += 1


    def tearDown(self):
        print(f"Test Results: {self.success}/{self.total_tests} passed, {self.failure}/{self.total_tests} failed")
        print(f"Success Rate: {(self.success/self.total_tests)*100:.2f}%")
        print(f"Failure Rate: {(self.failure/self.total_tests)*100:.2f}%")
        self.driver.quit()

login_test = LoginTest()
login_test.setUp()
login_test.test_valid_credentials()
login_test.test_invalid_credentials()
login_test.tearDown()
```
